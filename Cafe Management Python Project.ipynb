```python
from tabulate import tabulate
print("CAFE COFFEE DAY")
customerName = input("custmor name:")
contactNo = input("contact no:")



categories = [[1,"Beverages"],
              [2,"Shakes"],
              [3,"Lays and Chips"]
             ]

menu = [
    ["Tea", 10],
    ["Coffee", 20],
    ["Hot Coffee", 40],
    ["Cold Coffee", 50],
    ["cold drink",25]
]

menu2 = [
    ["Banana", 90],
    ["vanilla", 120],
    ["Mango", 80],
    ["Banana Khajur", 130],
    ["Chocolate",110],
    ["Strawberry",140],
    ["Chocolate banana",150],
    ["Cookies and Cream",120],
    ["Mocha",145],
    ["Nutella",120]
]

menu3 = [["Lays",10],
         ["Kurkure",10],
         ["Uncle chips",10]
        ]

head = ["S No","Categories"]
head2 = ["Beverages", "Price"]
head3 = ["Shakes", "Price"]
head4 = ["Chips","Price"]

print(tabulate(categories, headers = head, tablefmt = "rounded_grid"))
v = "Yes"

totalOrderedAmount = 0
orderitems = []
Q = input("Do you want some beverages(Yes/No):")
if Q == v:
    print(tabulate(menu, headers=head2, tablefmt="rounded_grid"))
    print("order from above menu")
    noOfItems = input('Please enter number of items:')
    for i in range(0, int(noOfItems)):
        c = input()
        orderitems.append(c)

    for i in orderitems:
        for j in menu:
            if i == j[0]:
                totalOrderedAmount = totalOrderedAmount + j[1]
    
W = input("Do you want some shakes(Yes/No):")
if W == v:
    print(tabulate(menu2, headers=head3, tablefmt="rounded_grid"))
    print("order from above menu")
    noOfItems = input('Please enter number of items:')
    for i in range(0, int(noOfItems)):
        c = input()
        orderitems.append(c)

    for i in orderitems:
        for j in menu2:
            if i == j[0]:
                totalOrderedAmount = totalOrderedAmount + j[1]
                    
E = input("Do you want some Lays(Yes/No):")
if E == v:
    print(tabulate(menu3, headers = head4, tablefmt = "rounded_grid"))
    print("order from above menu")
    noOfItems = input('Please enter number of items:')
    for i in range(0, int(noOfItems)):
        c = input()
        orderitems.append(c)

    for i in orderitems:
        for j in menu3:
            if i == j[0]:
                totalOrderedAmount = totalOrderedAmount + j[1]

print(customerName, ",your total bill is:", totalOrderedAmount)
print('Further information will be sent to your contact info:', contactNo)
```
